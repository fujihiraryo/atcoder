```python
N=int(input())
S_list=[]
for n in range(N):
    S_list.append(input())
    
S_list=sorted([sorted(s) for s in S_list])   
count_list=[]
tmp_count=0
for n in range(N-1):
    left=S_list[n]
    right=S_list[n+1]
    if left==right:
        tmp_count+=1
    else:
        count_list.append(tmp_count+1)
        tmp_count=0
count_list.append(tmp_count+1)
count_list=[c*(c-1)//2 for c in count_list]
print(sum(count_list))
```

```python
N,M=map(int,input().split())
work_list=[]
for n in range(N):
    A,B=map(int,input().split())
    work_list.append({'day':A,'price':B})
work_list.sort(key=lambda x:x['day'], reverse=True)
work_list.sort(key=lambda x:x['price'], reverse=True)

limit=M
money=0

for n in range(M+1):
    decision=False
    while not decision and len(work_list)!=0:
        work=work_list.pop(0)
        if work['day']<=limit:
            decision=True
            money+=work['price']
    limit-=1
print(money)
```
