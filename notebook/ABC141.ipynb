```python
S=input()
if S=="Sunny":
    print("Cloudy")
if S=="Cloudy":
    print("Rainy")
if S=="Rainy":
    print("Sunny")
```

```python
S=input()
n=len(S)
res="Yes"
for i in range(1,n+1):
    if i%2==1 and S[i-1]=="L":
        res="No"
    if i%2==0 and S[i-1]=="R":
        res="No"
print(res)
```

```python
import collections
N,K,Q=map(int,input().split())
A=[int(input()) for q in range(Q)]
c=collections.Counter(A)
for n in range(N):
    if K-Q+c[n+1]>0:
        print("Yes")
    else:
        print("No")
```

```python
from collections import deque
N,M=map(int,input().split())
A=[int(_) for _ in input().split()]
A.sort()
max_A=A[-1]
A=set(A[:-1])

for m in range(M):
    A.add(max_A//2)
    max_A=max(A)
    
    
print(sum(A))
```

```python
import heapq

N,M=map(int,input().split())
A=[-int(_) for _ in input().split()]
heapq.heapify(A)

for _ in range(M):
    a=heapq.heappop(A)
    heapq.heappush(A,-((-a)//2))

print(-sum(A))
```
